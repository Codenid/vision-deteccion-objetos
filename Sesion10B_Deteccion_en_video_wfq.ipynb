Importamos librerías

```python
from ultralytics import YOLO
from pathlib import Path
from IPython.display import HTML
from base64 import b64encode
import os
```

# (A) Configuración

```python
INPUT_VIDEO  = "Sesion10_video.mp4"        # ← ajusta si tu archivo tiene otro nombre
CONF_THRESH  = 0.50                         # confianza mínima
RUN_NAME     = "sesionCV_deteccion"                 # nombre de subcarpeta que se creará. Aquí se guardará el resultado (en runs/detect/RUN_NAME)
DEVICE       = "cpu"                        # 0 → GPU, "cpu" → CPU

assert Path(INPUT_VIDEO).exists(), f"No se encuentra {INPUT_VIDEO}. --> Revisa la ruta"
```

# (B) Cargar modelo y procesar video

```python
model = YOLO("yolo11m.pt")                    # modelo nano pre-entrenado COCO
results = model.predict(
    source      = INPUT_VIDEO,                # video de entrada
    conf        = CONF_THRESH,               # umbral de confianza
    device      = DEVICE,
    save        = True,                      # guarda un video anotado
    project     = "runs",                    # carpeta raíz
    name        = RUN_NAME,                  # subcarpeta específica
    vid_stride  = 1                         # procesa todos los frames
)

out_path = Path("runs") / RUN_NAME / Path(INPUT_VIDEO).name
print(f"Video procesado guardado en:\n{out_path}")
```
